```python
missing_data = td.isnull().sum()
print(missing_data)
```

```python
td['Age'].fillna(td['Age'].median(), inplace=True)
td['Cabin'].fillna('Unknown', inplace=True)
td['Embarked'].fillna(td['Embarked'].mode()[0], inplace=True)
```

```python
td.head(10)
```

```python
td.tail(10)
```

```python
td.drop(['PassengerId', 'Name', 'Ticket'],axis='columns', inplace=True)
td.head()
```

```python
td.head(10)
```

```python
td.tail(10)
```

```python
summary = td.describe()
print(summary)
```

```python
td.groupby('Sex').Survived.count()
```

```python
sns.countplot(data=td,x='Sex',hue='Survived');
plt.title('Count of Survived (0 = Male, 1 = Female)')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=td)
plt.title('Age by Passenger Class (Pclass)')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=td)
plt.title('Survival by Passenger Class (Pclass)')
plt.show()
```
